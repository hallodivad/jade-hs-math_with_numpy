```python
import numpy as np
```

```python
A=np.matrix([[1,2,1],[1,1,3],[1,4,2]])
print(A)
```

```python
b=np.matrix([[4],[10],[7]])
print(b)
```

```python
Ainv=np.linalg.inv(A)
print(Ainv)
```

```python
x=Ainv*b
print(x)
```

```python
x=np.linalg.solve(A,b)
print(x)
```

```python
print(A*x)
```
