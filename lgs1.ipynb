```python
import numpy as np
```

Nun können wir die Bibliothek numpy mit der Abkürzung np nutzen.
Wir definieren eine Matrix A und lassen Sie anzeigen

```python
A=np.matrix([[1,2,1],[1,1,3],[1,4,2]])
print(A)
```

nun einen Vektor b

```python
b=np.matrix([[4],[10],[7]])
print(b)
```

Wir wollen nun das LGS Ax=b loesen. Erste Moeglichkeit: inverse Matrix berechnen und mit dieser x berechnen:

```python
Ainv=np.linalg.inv(A)
print(Ainv)
```

```python
x=Ainv*b
print(x)
```

Sieht komisch aus, ist aber richtig. e-16 heiszt mal 10 hoch -16. Der zweite Eintrag ist also so gut wie 0. Bestimmt ist er sogar exakt 0 und der Rest ist Rundungsfehler. Probe:

```python
print(A*x)
```

Nun zweite Moeglichkeit mit dem solve Befehl. Dieser ist der aufwendigen Inversenberechnung meist vorzuziehen. Hier koennte sich z.B. das numerische Gaussverfahren hinter verstecken. Wahrscheinlich aber noch etwas cleverer, als wir es kennengelernt haben... 

```python
x=np.linalg.solve(A,b)
print(x)
```

```python
print(A*x)
```
