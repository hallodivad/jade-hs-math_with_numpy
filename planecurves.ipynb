## Ebene Kurven malen mit Numpy und Matplotlib

```python
# we will need this for showing animations:
%matplotlib notebook
# Import packages
import numpy as np
import matplotlib.pyplot as plt
import time
#import mpld3
```

```python
# create a discrete time array from t1 to t2 with time step dt
t1=0
t2=1
dt=.1

t=np.arange(t1,t2+dt,dt)
```

```python
#calculate corresponding x and y coordinates
#here you can use different functions to draw different curves

x=3*np.cos(2*np.pi*t)
y=3*np.sin(2*np.pi*t)
```

```python
#draw the points corresponding to all x and y coordinates
fig, ax=plt.subplots()
ax.plot(x,y,'.r')
ax.set_xlabel("x")
ax.set_ylabel("y")

# draw the corresponding polygonal line
ax.plot(x,y,'-b')
```

### Animated Motion

```python
fig, ax=plt.subplots()
ax.plot(x,y,'-c')
ax.set_xlabel("x")
ax.set_ylabel("y")
pos, =ax.plot(x[0],y[0],'or')

for n in range(len(t)):
    pos.set_data(x[n],y[n])
    fig.canvas.draw()
    fig.canvas.flush_events()
    time.sleep(.1)
    
#plt.close(fig)
```
