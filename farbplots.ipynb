```python
#import libraries
import numpy as np
import matplotlib.pyplot as plt

# Define Grid for xy-plane:
# Gitter fuer die xy-ebene definieren
# hier 1000 Werte in x- und in y-Richtung
x=np.linspace(-10,10,1000)
y=np.linspace(-10,10,1000)

X,Y= np.meshgrid(x,y)

##Calculate z-values
Z=9*np.exp(-(X**2+Y**2)/10)

##plotting
fig, ax=plt.subplots(figsize=(10,10))
img=ax.imshow(Z,extent=[-10, 10,-10,10])
cb=fig.colorbar(img)
cb.set_label("z",fontsize=20)
ax.set_xlabel("x",fontsize=20)
ax.set_ylabel("y",fontsize=20)
plt.show()
```
