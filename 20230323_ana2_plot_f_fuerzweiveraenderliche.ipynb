```python
#import libraries
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.ticker import LinearLocator
```

## Surface Plot

```python
# Gitter fuer die xy-ebene definieren
# hier 1000 Werte in x- und in y-Richtung
x=np.linspace(-10,10,1000)
y=np.linspace(-10,10,1000)

X,Y= np.meshgrid(x,y)

#Calculate corresponding z=f(x,y) values
Z=9*np.exp(-((X)**2+(Y)**2)/10)
#Z=np.cos(np.pi*X)*np.cos(np.pi*Y)*3


fig, ax = plt.subplots(subplot_kw={"projection": "3d"},figsize=(8,8))

# Plot the surface.
surf = ax.plot_surface(X, Y, Z, cmap=cm.viridis,
                       linewidth=0, antialiased=False)

# Plot settings
#ax.set_zlim(-1.01, 1.01)
ax.zaxis.set_major_locator(LinearLocator(10))
# A StrMethodFormatter is used automatically
ax.zaxis.set_major_formatter('{x:.02f}')
# Add a color bar which maps values to colors.
fig.colorbar(surf, shrink=0.5, aspect=5)

plt.show()
```

## Farbdarstellungen

```python
# Gitter fuer die xy-ebene definieren
# hier 1000 Werte in x- und in y-Richtung
x=np.linspace(-10,10,1000)
y=np.linspace(-10,10,1000)

X,Y= np.meshgrid(x,y)

#Calculate corresponding z=f(x,y) values
Z=9*np.exp(-((X)**2+(Y)**2)/10)
#Z=np.cos(np.pi*X)*np.cos(np.pi*Y)*3

##plotting
fig, ax=plt.subplots(figsize=(5,4))
img=ax.pcolormesh(x,y,Z,cmap='viridis',shading='auto')

##plot settings:
cb=fig.colorbar(img)
cb.set_label("z",fontsize=20)
ax.set_xlabel("x",fontsize=20)
ax.set_ylabel("y",fontsize=20)
ax.axis('equal')
#ax.set(xlim=(-2,2),ylim=(-2,2))

```

## Kontourlinien

```python
# Gitter fuer die xy-ebene definieren
# hier 1000 Werte in x- und in y-Richtung
x=np.linspace(-5,5,1000)
y=np.linspace(-5,5,1000)

X,Y= np.meshgrid(x,y)

#Calculate corresponding z=f(x,y) values
Z=9*np.exp(-((X)**2+(Y)**2)/10)
#Z=1000*np.exp(-(.025*X**2+.15*Y**2))

##plotting
#Anzahl Kontourlinien
Nk=5
#plot contour lines
fig, ax=plt.subplots(figsize=(7,5))
cp=ax.contour(X,Y,Z,Nk,linewidths=5,cmap="viridis")

#plot settings
ax.clabel(cp, inline=1, fontsize=16,fmt='%1.1f')
ax.set_xlabel("x",fontsize=20)
ax.set_ylabel("y",fontsize=20)
ax.axis('equal')
#cb=fig.colorbar(cp)
ax.plot(4,1.7,'.r',markersize=20)
plt.show()
```
