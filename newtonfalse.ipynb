```python
import numpy as np
```

Zuerst einmal definieren wir die Funktion deren Nullstelle wir finden wollen.

```python
def f(x):
    """
    Funktion, deren Nullstelle gesucht wird
    """
    return x-np.log(3*x)
```

Und noch die Ableitung. Wir koennten natuerlich auch numerisch mit Python ableiten, aber wir nehmen hier mal die per Hand ausgerechnete exakte Ableitungsfunktion.

```python
def dfdx(x):
    """
    Ableitungsfunktion
    """
    return 1-1/x
```

So nun die benoetigten Variablen festlegen

```python
#Startwert festlegen
xstart=1.9
#Genauigkeit/Verbesserung fürs Abbruchkriterium festlegen
Dxmin=.0001

#weitere Initialisierungen...
#...der Verbesserung
Dx=100
#...und eines Iterationszaehlers
n=0
```

Und nun Newtonverfahren im while loop starten :)

```python
#while loop fuerst Newtonverfahren
while Dxmin>np.abs(Dx):
    n=n+1
    xneu=xstart-f(xstart)/dfdx(xstart)
    Dx=xneu-xstart
    print("Nach ",n,"Iterationsschritt(en): xneu= ",xneu," Verbesserung Dx=", Dx)
    xstart=xneu
    input("Press Enter to continue.")
```

Und nochmal das Endergebnis ausgeben:

```python
print("")
print("Eine Nullstelle von f ist also ungefähr bei xneu=", xneu)
print("mit f(xneu)=",f(xneu))
```
