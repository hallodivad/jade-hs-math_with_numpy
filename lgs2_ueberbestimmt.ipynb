```python
import numpy as np
```

Matrix A und Vektor b definieren

```python
A=np.matrix([[1,1],[1,2],[1,3],[1,4],[1,5]])
print(A)
```

```python
b=np.matrix([[2.2],[3.4],[6.9],[7.7],[10.1]])
print(b)
```

Nun definieren eine neue Matrix C=A^T A und einen neuen Vektor d=A^T d.

```python
C=np.transpose(A)*A
print(C)
```

```python
d=np.transpose(A)*b
print(d)
```

 Dann sind die Normalengleichungen A^T A x=A^T b gegeben durch C x = d. Diesen sind loesbar wenn die Determinante ungleich 0 ist.

```python
np.linalg.det(C)
```

Passt. Loesen liefert dann

```python
x=np.linalg.solve(C,d)
print(x)
```

```python
print(C*x)
```

Probe stimmt, also Normalengleichungen sind geloest. Aber ist x nun auch eine ungfaehre moeglichst gute Loesung von A?

```python
print(A*x)
```

sieht doch ganz gut aus. ist aehnlich wie der b Vektor.
