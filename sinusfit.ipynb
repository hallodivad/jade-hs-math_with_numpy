```python
# Import packages
import numpy as np
import matplotlib.pyplot as plt
import time
from scipy.optimize import minimize
```

```python
#Daten laden
t=np.loadtxt('t-werte.txt')
s=np.loadtxt('s-werte.txt')

#Datenpunkte Plotten
fig, ax=plt.subplots(figsize=(12,5))
ax.plot(t,s,'.b')
ax.set_xlabel("t [ms]")
ax.set_ylabel("s [?]")
plt.show()
```

### In der folgenden Definition der Abweichungsquadratsumme ist ein Fehler. Finden Sie Ihn, um eine gute Ausgleichskurve zu erhalten

```python
#Abweichungsquadratsumme definieren
def S(A,f,phi,t,s):
    return np.sum((s+A*np.sin(2*np.pi*f*t+phi))**2)
```

```python
#Abweichungsquadratsumme minimieren
#Hier wird obige Funktion S minimiert... 
#Code muessen Sie hier nicht verstehen. Ich sage auch nochmal was in der VL dazu.
la=minimize(lambda x: S(x[0],x[1],x[2],t,s),[2.1,.4,.35])

#liefert die optimalen Parameter
A=la.x[0]
f=la.x[1]
phi=la.x[2]

print("Die ermittelten Parameter sind:")
print()
print("A=",A)
print()
print("f="+str(f)+" kHz" )
print()
print("phi=",phi)
print()

#Funktionswerte des Modells bestimmen
smodel=A*np.sin(2*np.pi*f*t+phi)
```

```python
#Plotten
fig, ax=plt.subplots(figsize=(12,5))
ax.plot(t,s,'.b')
ax.plot(t,smodel,'-r')
ax.set_xlabel("t [ms]")
ax.set_ylabel("s [?]")
plt.show()
```
